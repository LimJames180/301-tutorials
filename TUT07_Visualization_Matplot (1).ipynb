# To Find the trend

```python
# get 50 random data from-1 to 1
data = np.random.randn(50)
mean_pnt = data.mean()
# Non-cumulative
fig = plt.figure(figsize=( 12,8))
plt.plot(data)
plt.axhline(mean_pnt, c='r', linestyle ='--')
```

# Annotation Practice

```python
# Cumulative
plt.plot(data.cumsum())
```

```python
y1 = np.random.randn(50).cumsum()
y2 = np.random.randn(50).cumsum()
d ={'Stock1': y1, 'Stock2': y2}
c =['g','r']
x = np.linspace(0,50,50)
for key,y in d.items():
    plt.plot(x,y, linestyle = '--', marker= 'o', label= key, linewidth= 3)

# add ticks on x-axis
# xx-small, x-small, small, medium, large, x-large, xx-large, larger, smaller, None
xtick_loc=[0,10,20,30,40,50]
xtick_label= ['zero','ten','twenty','thirty','fourty','fifty']
plt.xticks(xtick_loc, xtick_label, rotation = 20, fontsize='large')
# add ticks on y-axis
ytick_loc=[-4,-2,0,2,4]
ytick_label= ['-4%','-2%','0%','2%','4%']
plt.yticks(ytick_loc, ytick_label, rotation = 20, fontsize='large')
# add labels on x and y axis
plt.xlabel('Time', fontsize=18)
plt.ylabel('Returns', fontsize= 18)
# Add a title
plt.title('Simulated Cumulative Stock Returns, Daily', fontsize= 25)
# Add a grid
plt.grid()
# Tighten the axis
plt.axis('tight')
# add a legend
# it is better to label each line before here, so check
plt.legend(loc='best',fontsize=8)
# add a text
plt.text(30, -2, 'The 30th daily return', family ='sans-serif', fontsize = 10, c= 'g')
#add an annotate
x_min = np.argmin(y1)
y_min = np.min(y1)
plt.annotate('Min',xy=(x_min+0.5, y_min+0.05),xytext=(x_min+5, y_min+0.5), 
             arrowprops = dict(facecolor='yellow', headwidth=8, width=4, headlength=10),
                               horizontalalignment='right', verticalalignment='bottom')
# plt.show()
plt.show()
# save figure
filename='TUTLINE01'
plt.savefig(filename)
```

# Work with a DataFrame

```python
import pandas as pd
```

```python
url = 'https://github.com/jiechengh/mgt301/blob/main/airbnb_sin_nyc.csv?raw=true'
airbnb = pd.read_csv(url, index_col=0)
airbnb.dropna(inplace=True)
airbnb.head()
```

### Scatter Plots

```python
fig, ax= plt.subplots()
airbnb.plot(kind='scatter', x = 'number_of_reviews', y = 'reviews_per_month',
            ax = ax, c ='orange', marker ='<', figsize =(10,8), xlim=(0,400),
           ylim=(0,20), title='number_of_reviews vs reviews_per_month')
```

### Histograms with mean and median values

```python
gs=gridspec.GridSpec(1,1)
fig, ax= plt.subplots(figsize=(10,8), facecolor = 'yellow')

airbnb['number_of_reviews'].plot(kind = 'hist', bins =15, ax= ax )

plt.xlabel('Number of Reviews')

mean = airbnb['number_of_reviews'].describe().loc['mean']
median= airbnb['number_of_reviews'].describe().loc['50%']
ax.axvline(mean, c= 'red', ls = '-.') 
ax.axvline(median, c= 'yellow', ls = '-.') 
```

# Line Chart

```python
airbnb_copy= airbnb.sort_values(by= 'last_review')
```

```python
airbnb['cumsum_number_reviews']=airbnb_copy.number_of_reviews.cumsum()
```

```python
gs = gridspec.GridSpec(1,1)
fig, ax= plt.subplots(figsize=(10,8), facecolor = 'white')

airbnb.sort_values(by='last_review').plot(x='last_review',y='cumsum_number_reviews', ax= ax)
```

### Box Plots

```python
airbnb[airbnb.number_of_reviews<= 
       airbnb.number_of_reviews.quantile(0.9)]['number_of_reviews'].plot(
    kind='box', subplots=True, layout=(1,1),
sharex=False, sharey=False, figsize=(10,10))
```

### Bar Chart

```python
airbnb_ng_rt= airbnb.groupby(['neighbourhood_group','room_type'])
price = airbnb_ng_rt[['price']].sum()
price = price.unstack()
price['total']= price['price'].sum(axis=1)
price = price.sort_values(by = 'total', ascending = False)
del price['total']
price.price.plot(kind ='bar', stacked =False, title = 'Total price for each room type based on Neighborhood')
```

```python
airbnb['neighbourhood_group'].value_counts().plot(kind='bar', color = 'orange', alpha = 0.5)
```

```python
# 4.
plt.clf()

gs = gridspec.GridSpec(1,1)
fig = plt.figure()
## 1st figure
ax1 = fig.add_subplot(gs[0,0])

airbnb['number_of_reviews'].plot(kind = 'hist', bins =15, alpha= 0.3, color='b', xlim=(0,300))
## 2nd figure
ax2=ax1.twinx()
airbnb['reviews_per_month'].plot(kind = 'hist', bins =15,alpha= 0.3, color='r')
```

```python
# 5. 
plt.clf()
gs = gridspec.GridSpec(2,2)
fig = plt.figure(figsize=(20,18))
ax1=fig.add_subplot(gs[0,0])
ax2=fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,:])

ax1.scatter(airbnb['number_of_reviews'], airbnb['reviews_per_month']  )
ax2.bar(height =airbnb['neighbourhood_group'].value_counts(), x= airbnb['neighbourhood_group'].unique() )
ax3.hist(airbnb['reviews_per_month'])

ax2.tick_params(axis='x', rotation=30)
```
